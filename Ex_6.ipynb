```python
import math
A=1
T=math.pi*2
tlist=[]
rlist=[]
for i in range (201):
    t=i*T/40
    tlist.append(t)
    r=A*math.sin((2*math.pi*t)/T)
    rlist.append(r)
    
#print(rlist,tlist)
%matplotlib inline 
%config InlineBackend.figure_format= 'svg'
import numpy as np 
import matplotlib.pyplot as plt 

plt.plot(tlist,rlist)
plt.ylabel('$\eta$')
plt.xlabel("Time")
plt.show()


```
